**Sentencia `return`**

- Hace que termine la ejecucin de la funcin cuando aparece y el programa contina por su flujo normal.
- Se puede utilizar para devolver un valor.

La sentencia return **es opcional**, puede devolver, o no, un valor y es posible que aparezca ms de una vez dentro de una misma funcin.

```python
# return que no devuelve ningn valor

def cuadrado_de_par(numero):
    if not numero % 2 == 0:
        return
    else:
        print(numero ** 2)
```

```python
cuadrado_de_par(8)
```

```python
cuadrado_de_par(3)
```

```python
# Varios return en una misma funcin
def es_par(numero):
    if numero % 2 == 0:
        return True
    else:
        return False

```

```python
es_par(2)
```

```python
es_par(5)
```

```python
# Devolver ms de un valor con return en Python
def cuadrado_y_cubo(numero):
    return numero ** 2, numero ** 3
```

```python
cuad, cubo = cuadrado_y_cubo(4)
cuad, cubo
```

```python
def tabla_del(numero):
    resultados = []
    for i in range(11):
        resultados.append(numero * i)
    return resultados
```

```python
res = tabla_del(3)
res
```

```python
# En Python una funcin siempre devuelve un valor!

def saludo(nombre):
    print(f'Hola {nombre}')

print(saludo('Juan'))

# Como puedes ver, el print que envuelve a la funcin saludo() devuelve None.
```

# Parmetros de las funciones

- **Paso por valor:** Un lenguaje de programacin que utiliza paso por valor de los argumentos, lo que realmente hace es copiar el valor de las variables en los respectivos parmetros. Cualquier modificacin del valor del parmetro, no afecta a la variable externa correspondiente.

- **Paso por referencia:** Un lenguaje de programacin que utiliza paso por referencia, lo que realmente hace es copiar en los parmetros la direccin de memoria de las variables que se usan como argumento. Esto implica que realmente hagan referencia al mismo objeto/elemento y cualquier modificacin del valor en el parmetro afectar a la variable externa correspondiente.


https://i.stack.imgur.com/8XAQ1.gif


> **En Python todo es un objeto.**
En Python realmente es que se pasa por valor la *referencia del objeto*  Qu implicaciones tiene esto? 
>> Bsicamente que si el tipo que se pasa como argumento es inmutable, cualquier modificacin en el valor del parmetro no afectar a la variable externa <br> pero, si es mutable (como una lista o diccionario), s se ver afectado por las modificaciones. 

```python
# esta es una funcin bsica de suma
def suma(a, b):
  return a + b
```

```python
result = suma(1, 2)
result
```

Se puedes definir valores predeterminados para los parmetros, de esa manera Python interpretar que el valor de ese parmetro es el predeterminado si no se proporciona ninguno.

```python
def suma(a, b=3):
  return a + b

result = suma(1)
result
```

Puedes pasar los parmetros en el orden que desees, utilizando el nombre del parmetro.

```python
result = suma(b=2, a=2)
result
```

Sin embargo, no es posible pasar un argumento de palabra clave antes que uno que no sea de palabra clave

```python
result = suma(3, b=2)
```

```python
result2 = suma(b=2, 3)
```

Las funciones tambin son objetos, por lo que puedes asignarlas a una variable y usar esa variable como una funcin.

```python
s = suma
result = s(1, 2)
result
```

**mbito y ciclo de vida de las variables**

En cualquier lenguaje de programacin de alto nivel, toda variable est definida dentro de un mbito. Esto es, los sitios en los que la variable tiene sentido y dnde se puede utilizar.
>Los parmetros y variables definidos dentro de una funcin tienen un mbito local. Por tanto, estos parmetros y variables no pueden ser utilizados fuera de la funcin porque no seran reconocidos.

El ciclo de vida de una variable determina el tiempo en que una variable permanece en memoria. 
>Una variable dentro de una funcin existe en memoria durante el tiempo en que est ejecutndose dicha funcin. Una vez que termina su ejecucin, sus variables y parmetros desaparecen de memoria y, por tanto, no pueden ser referenciados.

```python
def saludo(nombre):
    x = 10
    print(f'Hola {nombre}')
```
